```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
lista = []
for i in range(10):
    lista.append(i)
G = nx.Graph()
G.add_nodes_from(lista)
for j in lista:
    for i in lista:
        G.add_edge(i,j)

pos = nx.spring_layout(G)
plt.figure(figsize=(6, 6))
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=500, edge_color='gray')
plt.show()
```
